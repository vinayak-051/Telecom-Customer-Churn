Target Variable — Churn Flag

```python
df = pd.read_csv(r"C:\Users\Admin\Documents\cleaned_data.csv")
print(df.head(5))
```

```python
def create_churn_target(df, threshold=-0.20):
    # Step 1: Sort data by service provider, state, and date
    df_sorted = df.sort_values(['Service_provider', 'State', 'date'])
    
    # Step 2: Calculate month-over-month percentage change
    df_sorted['subscriber_change'] = df_sorted.groupby(['Service_provider', 'State'])['value'].pct_change()
    
    # Step 3: Create churn flag
    df_sorted['churn_flag'] = (df_sorted['subscriber_change'] < threshold).astype(int)
    
    return df_sorted
```

```python
df = create_churn_target(df)
print(df[['Service_provider', 'State', 'date', 'value', 'subscriber_change', 'churn_flag']].head(5))
```

**Advanced Feature Engineering**

Time-Series Features

```python
def create_temporal_features(df):
    # Lag features
    for lag in [1, 3, 6, 12]:
        df[f'subscribers_lag_{lag}'] = df.groupby(['Service_provider', 'State'])['value'].shift(lag)
    
    # Rolling mean
    for window in [3, 6, 12]:
        df[f'subscribers_ma_{window}'] = df.groupby(['Service_provider', 'State'])['value'].transform(lambda x: x.rolling(window).mean())
    
    # Year to year growth
    df['yoy_growth'] = df.groupby(['Service_provider', 'State'])['value'].pct_change(12)
    
    return df
```

```python
df = df.sort_values(['Service_provider', 'State', 'date'])
df = create_temporal_features(df)
```
